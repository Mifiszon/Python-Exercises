```python
import re
import csv
from clp3 import clp

with open("pap.txt", "r", encoding="utf-8") as file:
    text = file.read().lower()
    words = re.findall(r'\b\w+\b', text)

w_count = {}

for w in words:
    ids = clp(w)
    if not ids:
        continue
    id0 = ids[0]
    base = clp.bform(id0)
    label = clp.label(id0)
    key = (base, label[0])
    
    if key in w_count:
        w_count[key] += 1
    else:
        w_count[key] = 1

sorted_w_count = sorted(w_count.items(), key=lambda x: x[1], reverse=True)

with open("pap_clp.csv", "w", encoding="utf-8", newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Pozycja", "Forma podstawowa", "Liczba wystąpień", "Pierwsza litera etykiety"])
    
    for num, ((base, label_letter), count) in enumerate(sorted_w_count, start=1):
        writer.writerow([num, base, count, label_letter])
```
