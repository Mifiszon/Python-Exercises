```python
import re
import csv

with open("pap.txt", "r", encoding="utf-8") as file:
    text = file.read()
    text = text.lower()
    words = re.findall(r'\b\w+\b', text)

    w_count = {}
    for w in words:
        if w in w_count:
            w_count[w] += 1 
        else:
            w_count[w] = 1
    
    sorted_w_count = sorted(w_count.items(), key=lambda x: x[1], reverse=True)
    
    #for word, count in sorted_w_count:
    #    print(f"{word}: {count}")

    with open("pap.csv", "w", encoding="utf-8",  newline='') as csvfile:
        writer = csv.writer(csvfile)
        writer.writerow(["Pozycja", "Słowo", "Liczba wystąpień"])

        num = 1
        for word, count in sorted_w_count:
            writer.writerow([num, word, count])
            num += 1
```
